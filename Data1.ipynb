```python
from collections import deque

def input_graph():
    graph = {}
    n = int(input("Enter number of nodes: "))
    for i in range(n):
        node = input(f"Node {i+1}: ").strip()
        neigh = input(f"Neighbors of {node} (comma-separated): ").strip()
        graph[node] = neigh.replace(" ", "").split(",") if neigh else []
    return graph

def bfs(graph, start, goal):
    visited = []
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        node = path[-1]
        
        if node not in visited:
            visited.append(node)

            if node == goal:
                return visited, path

            for nb in graph[node]:
                queue.append(path + [nb])

    return visited, []   # if goal not reachable

graph = input_graph()
start = input("Enter start node: ").strip()
goal = input("Enter goal node: ").strip()

visited, path = bfs(graph, start, goal)

print("\nBFS Result")
print("Visited Order:", visited)
print("Path:", " -> ".join(path) if path else "No path found")
```

```python
def input_graph():
    graph = {}
    n = int(input("Enter number of nodes: "))
    for i in range(n):
        node = input(f"Node {i+1}: ").strip()
        neigh = input(f"Neighbors of {node} (comma-separated): ").strip()
        graph[node] = neigh.replace(" ", "").split(",") if neigh else []
    return graph


def dfs(graph, start, goal):
    visited = []
    stack = [[start]]

    while stack:
        path = stack.pop()
        node = path[-1]

        if node not in visited:
            visited.append(node)

            if node == goal:
                return visited, path

            # Add neighbors in reverse order for correct DFS
            for nb in reversed(graph.get(node, [])):
                stack.append(path + [nb])

    return visited, []   # if goal not reached


# ---- Main Program ----
graph = input_graph()
start = input("Enter start node: ").strip()
goal = input("Enter goal node: ").strip()

visited, path = dfs(graph, start, goal)

print("\n=== DFS ===")
print("Visited Order:", visited)
print("Path to Goal:", " -> ".join(path) if path else "No path found")
```

```python
def a_star(graph, h, start, goal):
    print("\nðŸ”¹ A* Search Started")
    open_list = [(start, [start], 0)]   # (node, path, g-cost)
    visited = set()

    while open_list:
        # Sort by f(n) = g(n) + h(n)
        open_list.sort(key=lambda x: x[2] + h[x[0]])
        current, path, cost = open_list.pop(0)

        print("Visiting:", current)

        if current == goal:
            print("\nðŸŽ¯ Goal Reached!")
            return path

        visited.add(current)

        for neighbor, edge_cost in graph.get(current, []):
            if neighbor not in visited:
                g_cost = cost + edge_cost
                open_list.append((neighbor, path + [neighbor], g_cost))

    print("\nâŒ No Path Found")
    return None


# ---------- Input Section ----------
graph = {}
n = int(input("Enter number of nodes: "))

for _ in range(n):
    node = input("\nEnter node: ").strip()
    neighbors = input(f"Enter neighbors of {node} as (name cost), separated by commas: ").strip()

    neighbor_list = []
    if neighbors:
        for pair in neighbors.split(','):
            name, cost = pair.strip().split()
            neighbor_list.append((name, int(cost)))

    graph[node] = neighbor_list

# Heuristic input
h = {}
print("\nEnter heuristic values:")
for node in graph:
    h[node] = int(input(f"Heuristic for {node}: "))

start = input("\nEnter start node: ").strip()
goal = input("Enter goal node: ").strip()

# ---------- Run A* ----------
result = a_star(graph, h, start, goal)

# ---------- Output ----------
print("\n=== Final Result ===")
print("Algorithm : A* Search")
if result:
    print("Path      :", " -> ".join(result))
else:
    print("No valid path found")
```

```python
def greedy_best_first_search(graph, h, start, goal):
    print("\nðŸ”¹ Greedy Best-First Search Started")
    
    open_list = [(start, [start])]   # (current_node, path)
    visited = set()

    while open_list:
        # Choose node with the smallest heuristic value
        open_list.sort(key=lambda x: h[x[0]])
        current, path = open_list.pop(0)

        print("Visiting:", current)

        if current == goal:
            print("\nðŸŽ¯ Goal Reached!")
            return path

        visited.add(current)

        for neighbor, _ in graph.get(current, []):
            if neighbor not in visited:
                open_list.append((neighbor, path + [neighbor]))

    print("\nâŒ No Path Found")
    return None


# ---------------- INPUT SECTION ----------------
graph = {}
n = int(input("Enter number of nodes: "))

for _ in range(n):
    node = input("\nEnter node: ").strip()
    neighbors = input(f"Enter neighbors of {node} as (name cost), separated by commas: ").strip()

    neighbor_list = []
    if neighbors:
        for pair in neighbors.split(','):
            name, cost = pair.strip().split()
            neighbor_list.append((name, int(cost)))
    graph[node] = neighbor_list

# Heuristic input
h = {}
print("\nEnter heuristic values:")
for node in graph:
    h[node] = int(input(f"Heuristic for {node}: "))

start = input("\nEnter start node: ").strip()
goal = input("Enter goal node: ").strip()

# ---------------- ALGORITHM EXECUTION ----------------
result = greedy_best_first_search(graph, h, start, goal)

# ---------------- FINAL OUTPUT ----------------
print("\n===== Final Result =====")
print("Algorithm : Greedy Best-First Search")
if result:
    print("Path      :", " -> ".join(result))
else:
    print("No valid path found")
```

```python
# Knowledge base: Rules in the expert system
rules = [
    {"if": {"fever", "cough", "fatigue"}, "then": "Flu"},
    {"if": {"fever", "cough", "shortness of breath"}, "then": "COVID-19"},
    {"if": {"headache", "nausea", "sensitivity to light"}, "then": "Migraine"},
    {"if": {"sore throat", "runny nose", "sneezing"}, "then": "Common Cold"},
    {"if": {"chest pain", "shortness of breath", "dizziness"}, "then": "Heart Attack"},
]

# Forward chaining inference engine
def diagnose(symptoms):
    matched_diseases = []
    symptoms_set = set(symptoms)

    # Check all rules (not just first match)
    for rule in rules:
        if rule["if"].issubset(symptoms_set):
            matched_diseases.append(rule["then"])
    return matched_diseases

# User interface
def main():
    print("=== Forward Chaining Medical Expert System ===")
    print("Enter your symptoms separated by commas (e.g., fever, cough):")

    user_input = input("Symptoms: ").lower()
    symptoms = [s.strip() for s in user_input.split(",")]

    diagnoses = diagnose(symptoms)

    if diagnoses:
        print("\nPossible diagnosis(es):")
        for disease in diagnoses:
            print(f"- {disease}")
    else:
        print("\nNo matching diagnosis found. Please consult a doctor.")

# Run program
if __name__ == "__main__":
    main()
```

```python
def print_board(board):
    print(f"\n {board[0]} | {board[1]} | {board[2]}")
    print("---+---+---")
    print(f" {board[3]} | {board[4]} | {board[5]}")
    print("---+---+---")
    print(f" {board[6]} | {board[7]} | {board[8]}\n")


def show_positions():
    print("Positions are numbered 1 to 9 as below:")
    print(" 1 | 2 | 3 ")
    print("---+---+---")
    print(" 4 | 5 | 6 ")
    print("---+---+---")
    print(" 7 | 8 | 9 \n")


def is_valid_move(board, move):
    return move.isdigit() and 1 <= int(move) <= 9 and board[int(move) - 1] == " "


def check_winner(board, player):
    win_conditions = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),  # rows
        (0, 3, 6), (1, 4, 7), (2, 5, 8),  # columns
        (0, 4, 8), (2, 4, 6)             # diagonals
    ]
    return any(board[i] == board[j] == board[k] == player for i, j, k in win_conditions)


def is_draw(board):
    return " " not in board


def make_ai_move(board):     # Simple AI: chooses first free place
    for i in range(9):
        if board[i] == " ":
            board[i] = "O"
            return i + 1    # return position (1â€“9)


def main():
    board = [" "] * 9
    print("Welcome to Tic-Tac-Toe!")
    print("You are X, AI is O\n")

    show_positions()
    print_board(board)

    while True:
        move = input("Enter your move (1â€“9): ")
        if not is_valid_move(board, move):
            print("Invalid move. Try again!")
            print_board(board)
            continue

        move = int(move) - 1
        board[move] = "X"
        print_board(board)

        if check_winner(board, "X"):
            print("ðŸŽ‰ You Win!")
            break
        if is_draw(board):
            print("ðŸ¤ It's a Draw!")
            break

        ai_pos = make_ai_move(board)
        print(f"AI placed O at position {ai_pos}")
        print_board(board)

        if check_winner(board, "O"):
            print("ðŸ’» AI Wins!")
            break
        if is_draw(board):
            print("ðŸ¤ It's a Draw!")
            break


if __name__ == "__main__":
    main()
```

```python
def check_winner(board, player):
    win_patterns = [
        (0,1,2), (3,4,5), (6,7,8),  # rows
        (0,3,6), (1,4,7), (2,5,8),  # columns
        (0,4,8), (2,4,6)            # diagonals
    ]
    return any(board[a] == board[b] == board[c] == player for a, b, c in win_patterns)


def minimax(board, is_maximizing):
    # Terminal state scoring
    if check_winner(board, "O"): return 1
    if check_winner(board, "X"): return -1
    if " " not in board: return 0

    if is_maximizing:
        best = -float("inf")
        for i in range(9):
            if board[i] == " ":
                board[i] = "O"
                best = max(best, minimax(board, False))
                board[i] = " "
        return best
    else:
        best = float("inf")
        for i in range(9):
            if board[i] == " ":
                board[i] = "X"
                best = min(best, minimax(board, True))
                board[i] = " "
        return best


# ðŸ” Display all moves with scores
def show_moves_with_scores(board):
    print("Move â†’ Minimax Score")
    for i in range(9):
        if board[i] == " ":
            board[i] = "O"     # Try AI move
            score = minimax(board, False)
            board[i] = " "     # Undo
            print(f"{i + 1} â†’ {score}")


# Example board state
board = [
    "X","O","X",
    " ","O"," ",
    " "," "," "
]

show_moves_with_scores(board)
```
